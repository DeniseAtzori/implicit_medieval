```python
def print_generation_example(row, rank=None):
    """
    Красиво печатает один пример генерации Task B.
    """

    if rank is not None:
        print("=" * 100)
        print(f"EXAMPLE #{rank}")
        print("=" * 100)
    else:
        print("=" * 100)

    print(f"ID: {row['id']}")
    print(f"Gold type: {row['gold_type']}")
    print(f"Predicted type: {row['predicted_type']}")
    print(f"Uncertainty: {row['uncertainty_level']}")
    print(f"Vote distribution: {row['vote_distribution']}")

    print("\nTWEET:")
    print(row["tweet_text"])

    print("\nGOLD REFERENCE:")
    print(row["gold_reference"])

    print("\nPREDICTED ARGUMENT:")
    print(row["predicted_argument"])

    print("\nROUGE-L:")
    print(f"canonical: {row['rouge_l_canonical']:.3f}")
    print(f"max majority refs: {row['rouge_l_max_majority_refs']:.3f}")
    print(f"max valid refs: {row['rouge_l_max_valid_refs']:.3f}")

    print("\nALL VALID REFERENCES:")
    refs = row["gold_all_valid_references"]

    if isinstance(refs, str):
        print(refs)
    else:
        for i, ref in enumerate(refs, start=1):
            print(f"{i}. {ref}")

    print("\n")
```

```python
best_examples = df.sort_values(
    by="rouge_l_max_valid_refs",
    ascending=False
).head(5)

for i, (_, row) in enumerate(best_examples.iterrows(), start=1):
    print_generation_example(row, rank=i)
```

```python
worst_examples = df.sort_values(
    by="rouge_l_max_valid_refs",
    ascending=True
).head(5)

for i, (_, row) in enumerate(worst_examples.iterrows(), start=1):
    print_generation_example(row, rank=i)
```
